## Some definitions
__Implicant__ = group of 1s  
__Prime implicant__ = largest group of 1s  
__Essential Prime implicant__ = Group of 1s that cover at least one minterm that can't be covered by another prime implicant.
Basically, a nonredundant group of 1s  
__Logical adjacent minterm__ is a minterm which differ in only one adjacent minterm.

# Quine-McCluskey tabular minimization method, or QM method
This allows you to use more vars, and is more systematic  

__STEPS:__  

1. List in a column all minterms of a function. Partition them into groups according to the number of 1/H bits.
This makes finding logically adjacent minterms easier.   
 
2. Search exhaustively between neighbouring groups for adjacent minterms and combine them into a column of (n-1)-variable  
implicants, checking off each minterm thats combined. The representation of each new implicant contains a dash in the  position of the eliminated variable. Repeatfor each column, combining (n-1)-variable implicants into (n-2)-variable   implicants and so on, untill no further implicants can be combined.  
Any term not checked off represents a prime implicant of the function, since its not covered by a larger implicant. The final result is a list of prime implicants of the function.  

3. Construct a prime implicant chart that lists minterms along the horizontal and prime implicants along the vertical, with an __x__ entry placed where a certain prime implicant covers a given minterm.  

4. Select a minimum number of prime implicants that cover all the minterms of the function.
