# BLZ 46, example 1.13
## plot this: 10 e^(-0.5t) * sin(4pi t) + 20 e^-t * sin(2pi t) 

```python
# voorbeel plot this: 10e−0.5tsin(4πt) + 20e−tsin(2πt)
t = arange(0,10,0.05)
# ^ numpy.arange is like python range, but for floats too, and returns a np.array already
xt = 10 * exp(-0.5 * t) * sin(4 * pi * t) + 20 * exp(-t) * sin(2 * pi * t)
plot(t, xt)
```
