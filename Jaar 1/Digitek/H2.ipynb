## 2.2 Switching functions

Switching functions represent functions but for switching algebra, and can be defined like this:

$f(X_1, X_2, \dots , X_n)$ is a switching function of $X_1, X_2, \dots , X_n$ and can have the value of 1 or 0 depending on the values of the variables.

## SOP and POS forms
SOP (sum of products) are constructed by summing product terms, where each product is formed by ANDing a number of complemented and uncomplemented vars. An example is $f(A, B, C, D) = A \C B C + B D + \C A C \C D$   
POS(product of sums) forms are constructed by taking the product of sum terms, where each sum term is formed by ORing vars. an example of a POS form of a function is
$f(A,B,C,D) = (\C A + B + C)(B + C + D) (A + \C C + D)$

# Canonical Forms
There are forms of SOP and POS functions with special characteristics.
We've seen that a function can be represented by many different but equivalent ways. The canonical POS and SOP forms
are unique for each function.  

## Minterms 
For a function of n vars, if a product term contains each of the n vars exactly one time in (un)complemented form, the product
is called a minterm. if a function is represented by a sum of minterms only, the function is siad to be _canonical SOP_. Here's one: 
$f_\alpha (A,B,C) = \C AB \C C + AB \C C + \C A B C + ABC$

To simplify writing the canonical SOP, we commonly use a special notation where each minterm is represented by n-bit binary code.  
Each bit represents one of the vars of the midterm like this:  
- uncomplemented var = 1    
- complemented var = 0   

The vars are listed in the same order in each minterm. The usefulness of this notation is that, __for a minterm to evaluate to 1, each uncomplemented var in the minterm must be 1__.

The order of the vars in functional notations is very important and should not change, lest the meaning of the function changes.

## Maxterms.
If a sum term of a function of n vars contain each of the vars exactly one time
in whatever form, the sum is called a maxterm. If a function is represented as a product of 
sum terms, each of wich is a max term, the function is a canonical POS form. 
For example, $f^\gamma (A,B,C) = (A+B+C)(A+B+\C C)(\C A + B + C) (\C A + B + \C C)$.
We adopt a sepecial notation like with minterms, but with one difference: They are swapped now.  
- Uncomplemented var = 0  
- complemented var = 1

The advantage of this is that, for a maxterm to evaluate to 0, each uncomplemented var in maxterm must be 0.

## 2.2.4 Incompletely specified functions
In the design of digital circuits, we often see cases where the switching function is not completely specified. In other words, 
a function may be required to contain certain minterms and omit others, with the remaining minterms being optional. In this case,
 the optional minterms may be included in the logic design if they help simplify circuit logic, or ommited otherwise. 
 optional minterm is called a _don't care minterm._ Same for maxterms. 
 
 These don't cares arise in 2 ways:  
 1. Certain input combos may never be applied to a particular switching network. Since they never occus, their minterms may be used in amy manner we choose. 
 An example would be  the minterms $m_{10} \dots m_{15}$ in BCD digits. 
 
2. Where all input combos occur, but the output is only required to be 1 or 0 for certain combos. When writing expressions. the don't-cares will be labeled $d_i$ instead of $m_i$, or $D_i$ instead of $M_i$

### 2.5.1 AND-OR and NAND networks
An AND-OR network employs AND gates to form product terms and OR gates to form the sum of these products. So, a function 
that is inplemented in AND-OR logic must be written in SOP form.



### 2.5.2 OR-AND and NOR networks

An OR-AND network employs OR to make sums and AND gates to form producs. Therefore, functions that need to be implemented in this form need to be written in POS form.



### 2.5.3 two-level circuits

Networks that have a structure like a circuit that combines 2port OR to create 4 port OR are referred to as two-level networks.
The inputs must pass throught 2 levels of gates before reaching output.
A network has __n__ levels when at least _one_ input signal must pass through __n__ gates before reaching the output.

SOP and POS forms can directly be implemented in a 2-level network if the inputs are available in both un/complemented form, and a 3-level network when only one form is available(addition of NOT gates).
Often, more levels are needed when there are gate _fan-in limits_ (e.g. you need a 3 input ANDs but are restricted to  2 input ANDs.)

### 2.5.5 Factoring
Factoring is a technique for obtaining higher level forms of switching functions. These have the benefit of requiring less hardware and thus being cheaper to implement. 
They're also needed in situations where gates with limited fan-in must be used. Here, factoring is used to reduce the number of literals in a large POS or SOP to  alues less than or equal to the available number of gate inputs. These are harder to design than simple POS or SOP forms, 
and are generally slower, due to having more levels of gates. 
Factoring is mostly dependend on the distributive law (Axiom 5) and is a bit of an artform where experience plays a big role. The technique is further complicated by the fact that redundancy may have to be added as a intermediate step in order to obtain a simpler function through factoring.
