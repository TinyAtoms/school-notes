```python
def response_trlc(t, R, L, C):
    # will write the equation in parts and then multiply, because it's getting hella hard to read
    tau = float(1/ (L * C))
    constant = float(R / (2 * L)) 
    vc_p1 = (tau - constant **2 ) ** -0.5
    vc_p2 = exp( -1 * t * constant)
    vc_p3 = sin(((tau - constant**2)** 0.5)*t)
    vc = vc_p1 * vc_p2 * vc_p3 
    return vc

t = arange(0, 12, 0.1)
vc1 = response_trlc(t, 1, 0.1, 1)
vc2 = response_trlc(t, 0.5, 1, 0.5)
vc3 = response_trlc(t, 0.5, 1, 0.1)

data  = array([t, vc1, vc2, vc3])
multiplot(data, ["critically dampened", "underdampened", "underdamp. + h.freq."])
```
