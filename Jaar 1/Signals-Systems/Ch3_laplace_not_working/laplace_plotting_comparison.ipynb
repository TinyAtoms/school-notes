```python
#these are the results obtained from book, octave and python
pythons = -2 * stepfunction(0) + 2 + (1/(s+3))
book = 2 + (1/(s+3))
print(array_equal(pythons, book))
plot(s, pythons)
```

```python
# results from book and python
pythons = ((-2 * s * exp(s) + s + 1) * exp(-s)) / (s**2 + s)
book = 1/s * exp(-s) - 2/(s+1)
print(sum(pythons- book), " is attributable to rounding error/working with floats")
multiplot([s, pythons, book], ["python","book"])
```
