```python
from sns import *
from numpy import *
from scipy import signal
import sympy as sym
```

```python
t = arange(0, 100, 0.1)
octave = -2 * exp(-t)
matlab = ((4 * pi * dirac(t)) - 2* pi * exp(-t) * (sign(t) + 1)) / (2 * pi)
book = ((4*pi*dirac(t))/exp(t) - (4*pi*stepfunction(t))/exp(t))/(2*pi)
```

```python
data = array([t, octave, matlab, book])
multiplot(data, ["octave", "matlab", "book"])
```

```python
data = array([t, octave, matlab])
multiplot(data, ["octave", "matlab"])
```

```python
data = array([t, octave])
multiplot(data, ["octave"])
```
