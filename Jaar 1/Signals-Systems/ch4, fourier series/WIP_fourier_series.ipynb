```python
N = 100
t = arange(0, 2, 0.01)
xN = full(200, 5, float) # creates an array with size 200 filled with 5, type float
factor = -40/ (pi * pi)
for n in range(1, N+1):
    xN += factor*cos((2 * n - 1) * pi * t) / ((2*n -1) ** 2)
plot(t, xN)
```

```python
t = arange(0, 6, 0.01)
N = 1000
xt = full(len(t), 0, float) # creates an array with size 200 filled with 0, type float
a0 = 2.5
for n in range(1, N+1):
    bn = 5/ (pi * n) # book says so, also says a = 0
    xt += a0 *  bn * sin(n * pi * t)
plot(t, xt)
```

```python
t = arange(-1,1, 0.01)
N = 5
xt = zeros(len(t))
for n in range(N+1):
    sinfill = (n * pi * t)/ 2
    an = (-1)**n / 10
    xt += an * sin(sinfill)
plot(t, xt)
```

```python
t = arange(0, 2, 0.01)
xt = 10 + 6 * cos(2 * pi * t) + 4 * cos(4 * pi * t)
plot(t, xt)
```

```python
t = arange(0, 3, 0.01)
N = 10
xt = zeros(len(t))
for n in range(1, N+1):
    an = 10 / sqrt((25 * n ** 2 * pi ** 2) + (2 * n ** 2 * pi ** 2 - 50)**2)
    phi = -1/tan((2 * n ** 2 * pi * pi - 50)/ (5 * n * pi))
    xt += an * sin(2 * n * pi * t + phi)
plot(t, xt)
```
